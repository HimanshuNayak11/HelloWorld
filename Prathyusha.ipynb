```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set(style="whitegrid")
```

```python
file_path = '/content/From Moo to YOU ! (Responses) (2).xlsx'
data = pd.read_excel(file_path)
```

```python
print("Data preview:")
display(data.head())
```

```python
print("\nData summary:")
data.info()
```

```python
print("\nDescriptive statistics for numeric columns:")
display(data.describe())
```

```python
print("\nDescriptive statistics for categorical columns:")
display(data.describe(include=['object', 'category']))
```

```python
features = {
    'Taste': 'What brand(s) would you choose based on the following features? [Taste]',
    'Packaging': 'What brand(s) would you choose based on the following features? [Packaging]',
    'Quality': 'What brand(s) would you choose based on the following features? [Quality]',
    'Aroma/Smell': 'What brand(s) would you choose based on the following features? [Aroma/Smell]',
    'Price': 'What brand(s) would you choose based on the following features? [Price]',
    'Visibility at store': 'What brand(s) would you choose based on the following features? [Visibility at Store]',
    'Trust': 'What brand(s) would you choose based on the following features? [Trust]'
}
```

```python
brand_colors = {
    'OMFED': 'pink',
    'Milky Moo': 'blue',
    'Pragati': 'red',
    'AMUL': 'green',
    'Mother Dairy': 'orange',
    'Loose': 'purple'
}

def get_color(value):
    return brand_colors.get(value, 'gray')
```

```python
def count_brands(series):
    brands = series.dropna().str.split(',').explode().str.strip()
    return brands.value_counts()
```

```python
for feature, column in features.items():
    brand_counts = count_brands(data[column])
    plt.figure(figsize=(10, 6))
    sns.barplot(x=brand_counts.values, y=brand_counts.index, palette=[get_color(brand) for brand in brand_counts.index])
    plt.title(f'Brand Preference Based on {feature}')
    plt.xlabel('Count')
    plt.ylabel('Brand')
    plt.show()
```

 Visualize Numeric Columns

```python
numeric_columns = data.select_dtypes(include=['number']).columns
for column in numeric_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

Visualize Categorical Columns

```python
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=data[column])
    plt.title(f'Count of each category in {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.show()
```

Analyze Correlations

```python
if len(numeric_columns) > 1:
    plt.figure(figsize=(12, 8))
    correlation_matrix = data[numeric_columns].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix')
    plt.show()
```

Check for Missing Values

```python
print("\nMissing values in each column:")
print(data.isnull().sum())
```

```python
data_filled = data.fillna(data.mean(numeric_only=True))
for column in categorical_columns:
    data_filled[column] = data_filled[column].fillna(data_filled[column].mode()[0])

print("\nRemaining missing values after handling:")
print(data_filled.isnull().sum())
```

```python
cleaned_file_path = 'cleaned_form_responses.xlsx'
data_filled.to_excel(cleaned_file_path, index=False)
print(f"\nCleaned data saved to {cleaned_file_path}")
```
